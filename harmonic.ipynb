```python
k=2.5
m=0.3
phi0=0
A=0.08
```

```python
import math
omega=math.sqrt(k/m)
```

```python
omega
```

```python
import numpy as np
t=np.linspace(0,7,101)
```

```python
t
```

```python
x=A*np.sin(phi0+omega*t)
```

```python
x
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(t,x,'g--')
plt.xlabel('t,seconds')
plt.ylabel('x,metres')
plt.title('harmonic notion')
        
```
